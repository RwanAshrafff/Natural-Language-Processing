```python
import requests
from collections import defaultdict, Counter
import random

url = "https://raw.githubusercontent.com/motazsaad/osac-corpus/master/0.7-1st_5000_transcript_clean.txt"
response = requests.get(url)
#response
corpus = response.text.splitlines()
corpus_lines =corpus[:5000]
corpus_lines
```

```python
tokens = []

for line in corpus_lines:
    words = line.split()   
    tokens.extend(words)
tokens
```

```python
vocab = set(tokens)
vocab_size = len(vocab)
print(vocab_size)
print(len(tokens))
```

```python
bigram_counts = {}
trigram_counts = {}

# Sliding through corpus to get bigram and trigram counts
for i in range(len(tokens) - 2):

    # generating a tuple for each consecutive words
    bigram = (tokens[i], tokens[i+1])
    trigram = (tokens[i], tokens[i+1], tokens[i+2])
    
    # Get the counts for bigram
    if bigram in bigram_counts.keys():
        bigram_counts[bigram] += 1
    else:
        bigram_counts[bigram] = 1
    
    # Get the counts for trigram
    if trigram in trigram_counts.keys():
        trigram_counts[trigram] += 1
    else:
        trigram_counts[trigram] = 1
```
